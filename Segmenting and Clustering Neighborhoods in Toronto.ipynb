# Segmenting and Clustering Neighborhoods in Toronto

## Scrape the following Wikipedia page

Using this tutorial: https://simpleanalytical.com/how-to-web-scrape-wikipedia-python-urllib-beautiful-soup-pandas

```python
# import the library we use to open URLs
import urllib.request
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup
```

```python
# specify which URL/web page we are going to be scraping
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
# open the url using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(url)
# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

Take a look at our underlying HTML code

```python
#print(soup.prettify())
```

```python
soup.title.string
```

Starting with an HTML < table > tag with a class identifier of "wikitable sortable". We’ll make a note of that for further use later.

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df = df.replace('\n','', regex=True)
df
```

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
index_drop = df[df['Borough'] == 'Not assigned'].index
df.drop(index_drop , inplace=True)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

Use the Geocoder package or the csv file to create the following dataframe:

```python
filename = "http://cocl.us/Geospatial_data"
#headers = ["Postal Code","Latitude","Longitude"]
df_postal = pd.read_csv(filename)
df_postal.set_index("Postal Code", inplace = True)
df_postal.head()
```

```python
#df['Postal Code'].unique()
```

```python
postal_data = df['Postal Code']
```

```python
df_postal.loc['M3A'][0]
```

```python
# define the dataframe columns
column_names = ['Latitude', 'Longitude'] 

# instantiate the dataframe
df_new = pd.DataFrame(columns=column_names)
```

```python
#Fill the dataframe with Latitude and Longitude
for postal in postal_data:
    df_new = df_new.append({'Latitude': df_postal.loc[postal][0],
                            'Longitude': df_postal.loc[postal][1]}, ignore_index=True)
```

```python
df['Latitude'] = df_new[Latitude']
df[Longitude] = df_new[Longitude]
```
